```python
import time, re, pandas as pd, winsound as sd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
# d1 turn on beep
def beepsound():
    fr = 2000
    du = 1000
    sd.Beep(fr, du)
try:  # except
    # 1 open file
    data = pd.read_csv(r"player_info.csv")
    # 2 crawling
    s = Service(r"C:\Users\NTX550\Desktop\chromedriver.exe")
    driver = webdriver.Chrome(service=s)
    driver.get(r"https://fifaonline4.nexon.com/datacenter")
    # 3 sort
    dtitle = {}
    for i in range(1, len(driver.find_elements(By.XPATH, """//*[@id="form1"]/div[2]/div[1]/div[2]/div[*]"""))+1): dtitle[driver.find_element(By.XPATH, """//*[@id="form1"]/div[2]/div[1]/div[2]/div[{0}]/label""".format(i)).get_attribute("title")] = i
    for i in range(9961, 10000): #여기 수정하면 됩니답
        # 6-2 initialization
        if i % 500 == 0:
            driver.close()
            driver = webdriver.Chrome(service=s)
            driver.get(r"https://fifaonline4.nexon.com/datacenter")
        # 4-1 normalization & calculation
        ntitle =re.sub('[)]', '', data.iat[i, 3].split('(')[1])
        title = dtitle[ntitle]
        keyword = data.iat[i, 2]
        driver.find_element(By.XPATH, """//*[@id="searchName"]""").clear()
        driver.implicitly_wait(10)
        driver.find_element(By.XPATH, """//*[@id="searchName"]""").send_keys(keyword)
        driver.implicitly_wait(10)
        elem = driver.find_element(By.XPATH, """//*[@id="form1"]/div[2]/div[1]/div[2]/div[{0}]""".format(title))
        driver.execute_script("arguments[0].scrollIntoView(true);", elem)
        elem.click()
        driver.implicitly_wait(10)
        driver.find_element(By.XPATH, """//*[@id="form1"]/div[1]/div[3]/button[1]""").click()
        driver.implicitly_wait(10)
        try : #exception handling
            elem2 = driver.find_element(By.XPATH, """//*[@id="divPlayerList"]/div/div[1]/div/div[6]/a""")
            driver.execute_script("arguments[0].scrollIntoView(true);", elem2)
            elem2.click()
            driver.implicitly_wait(10)
            time.sleep(2)
            temp = driver.find_element(By.XPATH, """//*[@id="playerPreview"]""").text.split('\n')
            rtemp = pd.DataFrame({"0":[temp[1]], "1":[temp[3]], "2":[ntitle]})
            n = 0
        except :
            temp = [data.iat[i,2]]
            temp.extend(["" for _ in range(60)])
            rtemp = pd.DataFrame({"0":[""], "1":[temp[0]], "2":[ntitle]})
            n = 1
        # 4-3 fit frame
        for i in range(26,30):
            if temp[i] == "골 결정력": n = i+1
        for i in range(n, n+11,2) : rtemp[i] = temp[i]
        rtemp[n+20] = temp[n+20]
        rtemp[n+22] = temp[n+22]
        rtemp[n+26] = temp[n+26]
        for i in range(n+30, n+51, 10) : rtemp[i] = temp[i]
        rtemp[n+52] = temp[n+52]
        # # 5 save file
        rtemp.to_csv(r"Player_Characteristic9961.csv", mode='a', header=False, index=False, encoding="utf-8-sig")
        # 6-1 initialization
        del rtemp
        elem = driver.find_element(By.XPATH, """//*[@id="form1"]/div[2]/div[1]/div[2]/div[{0}]""".format(title))
        driver.execute_script("arguments[0].scrollIntoView(true);", elem)
        elem.click()
        driver.implicitly_wait(10)
        time.sleep(1)
    beepsound()
except:
    pass
    beepsound()
```
